```python
def countMinCoins(amount, coin_map):
    coins = list(filter(lambda x: coin_map[x] > 0, list(coin_map.keys())))
    max = amount + 1
    dp = [float('inf')] * (max)
    dp[0] = 0
    
    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)
    return dp
    
def getDenominations(dp, amount, coins, coin_map, denomination=[]):
    if(amount == 0):
        return denomination
    for i in range(len(coins)):
        if(amount - coins[i] >=0 and dp[amount - coins[i]] + 1 == dp[amount]):
            denomination.append(coins[i])
            return getDenominations(dp, amount - coins[i], coins, coin_map, denomination)

amount_global = 10 # 10Kg
coin_map = {1: 15, 2: 5, 4: 7, 5: 2, 6: 3, 7: 2, 9: 4} # Weight - No.of packets map
# coin_map = {1: 1, 2: 1, 3: 1, 4: 1, 5: 1}

iter = 1
print(f"This is the order of optimal selection of packets to get a sum of {amount_global}")

# Just repeating the problem till all packet denominations become 0
while(len(list(filter(lambda x: coin_map[x] > 0, list(coin_map.keys())))) > 0):
    coins = list(filter(lambda x: coin_map[x] > 0, list(coin_map.keys())))
    dp_global = countMinCoins(amount_global, coin_map)
    denominations = getDenominations(dp_global, amount_global, coins, coin_map, [])
    print(f"Iter-{str(iter)} => {denominations}")
    for i in range(len(denominations)):
        if coin_map[denominations[i]] > 0:
            coin_map[denominations[i]] -= 1;
    iter+=1
```
